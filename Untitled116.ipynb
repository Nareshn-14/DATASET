# üìò 1. Import Required Libraries

```python
# Basic libraries
import numpy as np
import pandas as pd

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns

# For preprocessing and clustering
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

# üì• 2. Load and Inspect the Dataset

```python
# Load the CSV file
df = pd.read_csv('mall_customers.csv')

# Display the first few rows
df.head()
```

# üîç 3. Basic Data Exploration

```python
# Dataset info
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

# üßº 4. Data Preprocessing

```python
# Drop 'CustomerID' since it's not useful for clustering
df.drop('CustomerID', axis=1, inplace=True)

# Convert 'Gender' to numeric using Label Encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])  # Female=0, Male=1

# Standardize numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])

# Replace original columns with scaled values
df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = scaled_features

# Preview preprocessed data
df.head()
```

# üìä 5. Exploratory Data Analysis

# üîπ EDA 1: Gender Distribution (Count Plot)

```python
# Gender count plot
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

# üîπ EDA 2: Age Distribution (Histogram)

```python
# Age distribution histogram
plt.figure(figsize=(7, 4))
sns.histplot(df['Age'], bins=20, kde=True, color='teal')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# üîπ EDA 3: Annual Income vs Spending Score (Scatter Plot)

```python
# Income vs Spending scatter
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, hue='Gender', palette='Set2')
plt.title('Annual Income vs. Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```

# üî• EDA 5: Correlation Heatmap

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset (if not already loaded)
df = pd.read_csv('mall_customers.csv')

# Optional: Convert Gender to numerical if not done
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})

# Compute correlation matrix
corr_matrix = df[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

# üî¢ 6. Elbow Method to Determine Optimal k

```python
wcss = []  # Within-Cluster Sum of Squares
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K, wcss, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

# üß™ 7. Silhouette Score Calculation

```python
# Try for a few values of k to get silhouette scores
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df)
    score = silhouette_score(df, labels)
    print(f"Silhouette Score for k={k}: {score:.3f}")
```

# üß© 8. Apply K-Means Clustering


```python
# Final model with chosen number of clusters
kmeans_final = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans_final.fit_predict(df)
```

# üìà 9. Visualize Clusters

```python
# 2D scatter plot using the two most informative features
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'], 
                hue=df['Cluster'], palette='Set1', s=100)
plt.title('Customer Segments')
plt.xlabel('Annual Income (Standardized)')
plt.ylabel('Spending Score (Standardized)')
plt.grid(True)
plt.show()
```

# 10. Inverse Transform for Interpretation

```python
# Reverse standardization for income and spending score
df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = scaler.inverse_transform(
    df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
)

# Now you can interpret actual cluster centers
df.groupby('Cluster').mean()
```

# Save Clustered Results to CSV

```python
# Save the dataframe with cluster labels to CSV
df.to_csv('mall_customers_clustered.csv', index=False)

print("Clustered data has been saved to 'mall_customers_clustered.csv'")
```

# Plot Cluster Centers (for 2D visual)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Re-run clustering (if needed)
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Plot clusters and centers
plt.figure(figsize=(8, 6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], 
            c=df['Cluster'], cmap='rainbow', s=50, alpha=0.6, label='Customers')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            color='black', s=200, marker='X', label='Centroids')
plt.title('Customer Segments with Cluster Centers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```
